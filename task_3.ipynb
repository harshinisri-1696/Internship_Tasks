```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df=pd.read_csv('Iris.csv')
df
```

HISTORGAM OF VISUALIZE EACH COLUMN IN THE DATASET

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.histplot(df['SepalLengthCm'], kde=True, color="skyblue", ax=axes[0])
axes[0].set_title('Histogram of Sepal Length')
axes[0].set_xlabel('SepalLengthCm')
axes[0].set_ylabel('Frequency')

sns.histplot(df['SepalWidthCm'], kde=True, color="lightgreen", ax=axes[1])
axes[1].set_title('Histogram of Sepal Width')
axes[1].set_xlabel('SepalWidthCm')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.histplot(df['PetalLengthCm'], kde=True, color="skyblue", ax=axes[0])
axes[0].set_title('Histogram of Petal Length')
axes[0].set_xlabel('PetalLengthCm')
axes[0].set_ylabel('Frequency')

sns.histplot(df['PetalWidthCm'], kde=True, color="lightgreen", ax=axes[1])
axes[1].set_title('Histogram of Petal Width')
axes[1].set_xlabel('PetalWidthCm')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
plt.Figure(figsize=(8,6))
sns.histplot(df['Species'],kde=True, color="skyblue")
plt.title('HISTOGRAM OF SPECIES')
plt.xlabel('Species')
plt.ylabel("Frequency")
plt.show()
```

BAR GRAPH VISUALIZE OF EACH COLOUM IN THE DATASET

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

df["SepalLengthCm"].value_counts().sort_index().plot(kind="bar", color="yellow", ax=axes[0])
axes[0].set_title("Bar Graph of Sepal Length")
axes[0].set_xlabel("SepalLengthCm")
axes[0].set_ylabel("Frequency")
axes[0].tick_params(axis='x', rotation=45)

df["SepalWidthCm"].value_counts().sort_index().plot(kind="bar", color="red", ax=axes[1])
axes[1].set_title("Bar Graph of Sepal Width")
axes[1].set_xlabel("SepalWidthCm")
axes[1].set_ylabel("Frequency")
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

df["PetalLengthCm"].value_counts().sort_index().plot(kind="bar", color="red", ax=axes[0])
axes[0].set_title("Bar Graph of Petal Length")
axes[0].set_xlabel("PetalLengthCm")
axes[0].set_ylabel("Frequency")
axes[0].tick_params(axis='x', rotation=45)

df["PetalWidthCm"].value_counts().sort_index().plot(kind="bar", color="yellow", ax=axes[1])
axes[1].set_title("Bar Graph of Petal Width")
axes[1].set_xlabel("PetalWidthCm")
axes[1].set_ylabel("Frequency")
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
plt.Figure(figsize=(8,6))
df["Species"].value_counts().plot(kind="bar",color="pink")
plt.title(f"BAR GRAPH OF SPECIES")
plt.xlabel('Species')
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.show()
```

SCATTERPLOT VISUALIZE OF DATA IN THE DATASET
